## Importing the libraries

We would be using numpy and pandas

```python
import numpy as np 
import pandas as pd 
```

## Reading the data file

```python
data=pd.read_csv('global_temperature_anomalies.csv')
data
```

## Getting the column names

```python
data.columns
```

## Dropping the Non-Required Columns

```python
data.drop(['j_d', 'd_n', 'djf', 'mam', 'jja','son'],axis=1,inplace=True)
```

```python
data.head()
```

## Renaming the Columns

```python
columns=['Hemisphere', 'Year', 'January', 'February', 'March', 'April', 'May', 'June', 'July',
       'August', 'September', 'October', 'November', 'December']
data.columns=columns
data.head()
```

## Changing the data from wide to long format

```python
df=pd.melt(data,id_vars=['Hemisphere','Year'],value_vars=['January', 'February', 'March', 'April', 'May', 'June', 'July',
       'August', 'September', 'October', 'November', 'December'],value_name='Temperature Variation')
```

```python
df
```

## Adding a column of day

```python
df['day']=['01']*len(df.Year)
df.head()
```

## Checking the datatypes

```python
df.dtypes
```

## Converting datatype of Year from int to object

```python
df['Year']=df['Year'].astype(str)
df['Year'].dtype
```

```python
df.dtypes
```

## Adding a column of date

```python
df['Date']=df['day']+'/'+df['variable']+'/'+df['Year']
df.head()
```

## Checking the column names

```python
df.columns
```

## Rearranging the column names

```python
columns=['Hemisphere','day','variable' ,'Year' ,
       'Date','Temperature Variation']
df=df[columns]
df
```

## Chaning Column names

```python
columns=['Hemisphere','Day','Month' ,'Year' ,
       'Date','Temperature Variation']
df.columns=columns
df
```

## Saving to transformed

```python
df.to_csv('transformed_data.csv')
```
